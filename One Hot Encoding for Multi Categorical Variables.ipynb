```python
top_10 = [x for x in cat_col.X2.value_counts().sort_values(ascending=False).head(10).index]
```

```python
#Creating 10 binary columns

for label in top_10:
    cat_col[label] = np.where(cat_col['X2']==label,1,0)

cat_col[['X2']+top_10].head()
```

# Now Creating the function for getting the top 10 variable from each column

```python
def one_hot_encoding(df, columns, top_label):
    for label in top_label:
        df[columns+'_'+label] = np.where(df[columns]==label, 1,0)
        
```

```python
df = pd.read_csv('Mercedes-Benz.csv')

#taking the catogorical column
df1 = df.select_dtypes(object)
df1.head()
```

```python
x0_top_10 = [x for x in df1.X0.value_counts().sort_values(ascending=False).head(10).index]
x1_top_10 = [x for x in df1.X1.value_counts().sort_values(ascending=False).head(10).index]
x2_top_10 = [x for x in df1.X2.value_counts().sort_values(ascending=False).head(10).index]
x3_top_10 = [x for x in df1.X3.value_counts().sort_values(ascending=False).head(10).index]
x4_top_10 = [x for x in df1.X4.value_counts().sort_values(ascending=False).head(10).index]
x5_top_10 = [x for x in df1.X5.value_counts().sort_values(ascending=False).head(10).index]
x6_top_10 = [x for x in df1.X6.value_counts().sort_values(ascending=False).head(10).index]
x8_top_10 = [x for x in df1.X8.value_counts().sort_values(ascending=False).head(10).index]
```

```python
one_hot_encoding(df1,'X0', x0_top_10)
one_hot_encoding(df1,'X1', x1_top_10)
one_hot_encoding(df1,'X2', x2_top_10)
one_hot_encoding(df1,'X3', x3_top_10)
one_hot_encoding(df1,'X4', x4_top_10)
one_hot_encoding(df1,'X5', x5_top_10)
one_hot_encoding(df1,'X6', x6_top_10)
one_hot_encoding(df1,'X8', x8_top_10)
```

# Advantages of  One Hot Encoding of Top Variable
- straight forward to implement
- does not require hrs of variable exploration.
- Does not expand massivley the feature space(number of column in dataset.)

# Disadvantage of One Hot Encoding 
- does not add any information that make the varibale more predictve
- does not keep the information of the ignored label.
